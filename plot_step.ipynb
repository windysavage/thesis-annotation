```python
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
def get_train_log():
    train_acc = np.load('exp/step/history/train-acc.npy')
    train_loss = np.load('exp/step/history/train_loss.npy')
    return train_acc,train_loss
```

```python
def get_val_log():
    val_acc = np.load('exp/step/history/val-acc.npy')
    val_loss = np.load('exp/step/history/val_loss.npy')
    return val_acc,val_loss
```

```python
def plot_figure():
    train_acc,train_loss = get_train_log()
    val_acc,val_loss = get_val_log()
    
    plt.figure(figsize=(16,4))
    plt.subplot(121)
    plt.plot(range(target_epoch),train_loss[:target_epoch],label='train_loss')
    plt.plot(range(target_epoch),val_loss[:target_epoch],label='val_loss')
    #plt.xlim((0, 500))
    #plt.ylim((0, 4))
    plt.legend(loc='upper right')

    plt.subplot(122)
    plt.plot(range(target_epoch),train_acc[:target_epoch],label='train_acc')
    plt.plot(range(target_epoch),val_acc[:target_epoch],label='val_acc')

    plt.legend(loc='upper right')
    
    #plt.savefig('lr=0.00005&LSTM.png')
    plt.show()
```

```python
target_epoch = 268
plot_figure()
```
